```python
import pandas as pd
import numpy as np
from objects import Employee,Ticket,Solution,Role,Status
import db
import os
```

```python
# import data
df = pd.read_csv("../data/VectaCorp_DF.csv")
df.head(5)
```

```python
 # las tablas mas faciles de crear (ie, que no tengan llaves foraneas)
# role, solution, status
role_list = df['role'].unique()
status_list = df['status'].unique()
solution_list = df['solution'].unique()

print(role_list)
print(status_list)
print(solution_list)

# abrir conexion
db.connect()

# añadir roles
for element in role_list:
    role = Role(role=element)
    db.add_role(role)

# añadir solution
for element in solution_list:
    solution = Solution(solution=element)
    db.add_solution(solution)

# añadir status
for element in status_list:
    status = Status(status=element)
    db.add_status(status)

# añadir empleados
employee_list = list(zip(df['name'],df['username'],df['password'],df['email'],df['role']))
## valores unicos en una lista de tuplas
def removeDuplicates(lst):
    return [t for t in (set(tuple(i) for i in lst))]
employee_list = removeDuplicates(employee_list)

for element in employee_list:
    employee = Employee(name=element[0],
                        username=element[1],
                        password=element[2],
                        email=element[3],
                        roleid=db.get_employeeid_from_role(element[4]))
    db.add_employee(employee)

# añadir tickets
ticket_list = list(zip(df['customername'],df['customeremail'],df['submitteddate'],
                       df['issue'],df['username'],df['solution'],df['status']))
for element in ticket_list:
    ticket = Ticket(customername=element[0],
                    customeremail=element[1],
                    submitteddate=element[2],
                    issue=element[3],
                    employeeid=db.get_employeeid_from_username(element[4]),
                    solutionid=db.get_solutionid_from_solution(element[5]),
                    statusid=db.get_statusid_from_status(element[6]))
    db.add_tickets(ticket)



db.close()
```
